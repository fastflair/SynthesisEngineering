```python
import yt_dlp
import os

from spleeter.separator import Separator
from pydub import AudioSegment
```

```python
def download_audio_yt_dlp(url, output_path='output_audio.mp3'):
    ydl_opts = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'outtmpl': output_path,
        'noplaylist': True,
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        print(f"Audio successfully downloaded and saved to {output_path}")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
# Example usage
link = 'https://www.youtube.com/watch?v=gfHXG9IsulY'
output_file = 'flower'
download_audio_yt_dlp(link, output_path=output_file)
```

```python
# Create a separator with the 2stems configuration (vocals + accompaniment)
separator = Separator('spleeter:2stems')

# Define the path to the input audio file
input_audio = f'{output_file}.mp3'

# Separate the input audio into vocals and accompaniment
separator.separate_to_file(input_audio, './output_directory')

# Load the vocals stem that was generated by Spleeter
vocals_path = f'./output_directory/{output_file}/vocals.wav'
vocals = AudioSegment.from_wav(vocals_path)

# Save the vocals (speech) as a new audio file
output_speech_path = f'./{output_file}_speech_only.wav'
vocals.export(output_speech_path, format="wav")

print(f"Speech-only audio saved to: {output_speech_path}")
```
